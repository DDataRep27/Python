```python
# 4. Write a Python program to calculate the sum of three given numbers, if the values are equal then return three times of their sum hint: write User defined functions

x = int(input("Enter the 1st number: "))
y = int(input("Enter the 2nd number: "))
z = int(input("Enter the 3rd number: "))

def sum_thrice(x,y,z):
    sum = x+y+z
    if x==y==z:
        sum = sum*3
    return sum
print(sum_thrice(x,y,z))
```

```python
# 8. Write a Python program to get the Fibonacci series between 0 to 50.

# Note : The Fibonacci Sequence is the series of numbers :

# 0, 1, 1, 2, 3, 5, 8, 13, 21, ....

# Every next number is found by adding up the two numbers before it.


k,v = 0,1

while v<50 :
    print(v, end=" ")
    k, v = v, k+v
```

```python
# Write Python Programs to use various operators in Python

a = int(input("Enter the 1st number: "))
x = input("Operator: ")
b = int(input("Enter the 2nd number: "))

if x == ('*'):
    print("Solution:", a*b)
elif x==('+'):
    print("Solution:", a+b)
elif x==('-'):
    print("Solution:", a-b)
elif x==('/'):
    print("Solution:", a/b)
elif x==('>'):
    print("Solution:", a>b)
elif x==('<'):
    print("Solution:", a<b)
elif x==('//'):
    print("Solution:", a//b)
elif x==('%'):
    print("Solution:", a%b)
else: print("Operation cannot be performed, Please select following operators from list: '*', '+', '-', '>', '<', '//', '%' ")
```
